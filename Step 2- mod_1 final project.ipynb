```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import seaborn as sns
%matplotlib inline
```

```python
df_f = pd.read_csv('imdb_ahsan.csv')
```

```python
df_f.head()
```

```python
df = df_f.drop(columns = ['Unnamed: 0'])
```

```python
df.head()
```

## STEP 1 : EDA or DATA VISUALIZATION

### What is the relationship between genre and world wide gross?

```python
#scatterplot of genres vs. worldwide_gross (average)
sns.set(font_scale = 5)
sns.set(style="whitegrid", color_codes=True)
pal = sns.color_palette("Greens_d", len(df))
fig, ax = plt.subplots()
fig.set_size_inches(30,15)
ax = sns.barplot(x="genre", y="worldwide_gross", ci=None, palette=np.array(pal[::-1]),data=df)
#ax.set(xlabel='Movie Genre', ylabel='Average Movie Gross Revenue (U.S. Dollars)')
plt.xlabel('Movie Genres', fontsize=25, weight = 'bold')
plt.ylabel('Average Movie Gross Revenue (In Millions Of U.S. Dollars)', fontsize=25, weight = 'bold')
plt.yticks(ax.get_yticks(), ax.get_yticks()/1000000)
plt.tick_params(axis='both', which='major', labelsize=20)
#sns.set_style('darkgrid')
plt.show()
```

```python

#scatterplot of vote_count vs. vote_average
fig, ax = plt.subplots()
fig.set_size_inches(8,12)

ax = sns.scatterplot(x="vote_average", y="vote_count", marker="o", ci=68, data=df)
ax.xaxis.set_major_locator(ticker.MultipleLocator(1))
ax.xaxis.set_major_formatter(ticker.ScalarFormatter())
ax.yaxis.set_major_locator(ticker.MultipleLocator(1000))
ax.yaxis.set_major_formatter(ticker.ScalarFormatter())
ax.set(xlabel='Average Movie Rating (1-10)', ylabel='Number Of Votes')
sns.set(style='whitegrid', font_scale=1.2)
plt.show()
```

```python
#scatterplot of popularity vs. worldwide_gross
fig, ax = plt.subplots()
fig.set_size_inches(18,12)

ax = sns.scatterplot(x="popularity", y="worldwide_gross", marker="o", ci=68, data=df)
ax.xaxis.set_major_locator(ticker.MultipleLocator(10))
ax.xaxis.set_major_formatter(ticker.ScalarFormatter())
ax.set(xlabel='Movie Popularity (1-100)', ylabel='Movie Gross Revenue (U.S. Dollars)')
sns.set(style='darkgrid', font_scale=1.5)
plt.show()
```
